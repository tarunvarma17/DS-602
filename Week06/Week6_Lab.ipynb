```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/log_reg/employee-turnover-balanced.csv')
```

```python
cat_vars = list(df.dtypes[df.dtypes == object].index)
num_vars = list(df.dtypes[df.dtypes == 'int64'].index)
```

```python
X = df.drop(columns='left_company')
y = df['left_company']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=124)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler


num_pipeline = Pipeline([('impute_missing', SimpleImputer(strategy='median')),
                           ('standardize_num', StandardScaler())
                        ])

cat_pipeline = Pipeline([('impute_missing_cats', SimpleImputer(strategy='most_frequent')),
                          ('create_dummies_cats', OneHotEncoder(handle_unknown='ignore', drop='first'))])

processing_pipeline = ColumnTransformer(transformers=[('proc_numeric', num_pipeline, num_vars)])
```

```python
processing_pipeline = ColumnTransformer(transformers=[('create_dummies', cat_pipeline, cat_vars)])
```

```python
processing_pipeline = ColumnTransformer(transformers=[('proc_numeric', num_pipeline, num_vars),
                                                      ('create_dummies', cat_pipeline, cat_vars[1:])])
```

```python
from sklearn.linear_model import LogisticRegression

modeling_pipeline = Pipeline([('data_processing', processing_pipeline), 
                              ('lm', LogisticRegression())
                             ])
modeling_pipeline.fit(X_train, y_train)
```

```python
test_predictions = modeling_pipeline.predict(X_test)
train_predictions = modeling_pipeline.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print('''Test data accuracy (Before tuning) :''' + str(accuracy_score(y_test, test_predictions)))
```

```python
print('''Train data accuracy (Before tuning) :''' + str(accuracy_score(y_train, train_predictions)))
```

### After hyper-parameter tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'lm__penalty':["l1", "l2"],
              'lm__C': np.logspace(0,2,10),
              'lm__solver':['lbfgs', 'liblinear', 'newton-cg','saga']}

tuned = GridSearchCV(estimator = modeling_pipeline, param_grid = param_grid, scoring='accuracy')
```

```python
tuned.fit(X_train,y_train)
```

```python
tuned.best_params_
```

```python
test_predictions_tuned = tuned.predict(X_test)
train_predictions_tuned = tuned.predict(X_train)
```

```python
print('''Test data accuracy (After tuning) :''' + str(accuracy_score(y_test, test_predictions_tuned)))
```

```python
print('''Train data accuracy (After tuning) :''' + str(accuracy_score(y_train, train_predictions_tuned)))
```

### After tuning, we can see that the test score increased slightly, and the train score reduced. This is a success, as it shows that we have managed to reduce a bit of overfitting while also improving test score
